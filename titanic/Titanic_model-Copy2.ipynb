```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
import seaborn as sns
```

```python
titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')
```

```python
titanic_test.head(10)
```

```python
titanic_train.head(20)
```

```python
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train['Cabin'] = titanic_train['Cabin'].notnull().astype('int')
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train.head()
```

```python
titanic_train['Embarked'] = titanic_train['Embarked'].fillna('S')
titanic_train['Embarked'] = titanic_train['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
titanic_train = pd.get_dummies(titanic_train, columns=["Sex"])
titanic_train.head()
```

```python
titanic_train = titanic_train.dropna(how='any')
titanic_train.head()
len(titanic_train)
```

```python
X_final = titanic_train.drop(["Survived", "Ticket", "Name"], axis=1)
y_final = titanic_train["Survived"].values.reshape(-1, 1)
print(X_final.shape, y_final.shape)
```

```python
cor_map = titanic_train.drop(["Ticket", "Name", "PassengerId"], axis=1)
cor_map
```

```python
len(y_final)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, random_state=1, stratify=y_final)
```

```python
# X_final
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [10],
              'gamma': [0.0001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test)
```

```python
len(predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["Non-Survivors", "Survivors"]))
```

```python
import pickle

# Save to file in the current working directory
pkl_filename = "titanic_model_log.pkl"
with open(pkl_filename, 'wb') as file:
   pickle.dump(grid, file)

# Load from file
with open(pkl_filename, 'rb') as file:
   pickle_model = pickle.load(file)

# Calculate the accuracy score and predict target values
score = pickle_model.score(X_test, y_test)
print("Test score: {0:.2f} %".format(100 * score))
y_predict = pickle_model.predict(X_test)
```

```python
pickle_model.cv_results_
```

```python
from sklearn.feature_selection import SelectKBest, chi2
relevant_features_X_train = X_train.drop(labels=['Fare', 'PassengerId'], axis=1)
select_k_best = SelectKBest(score_func=chi2, k=len(relevant_features_X_train.columns))
select_k_best.fit(relevant_features_X_train, y_train)
(pd.DataFrame({'scores': select_k_best.scores_, 'features': relevant_features_X_train.columns })
    .sort_values(by='scores', ascending=False)
    .plot.bar(x='features', color = '#20b2aa', legend = None))

plt.tight_layout()
plt.savefig('features_graph.png')
```
