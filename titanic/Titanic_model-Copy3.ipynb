```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')
```

```python
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train['Cabin'] = titanic_train['Cabin'].notnull().astype('int')
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train.head()
```

```python
# df_with_dummies = pd.get_dummies( titanic train, columns = cols_to_transform )
```

```python
titanic_train = pd.get_dummies(titanic_train.drop(labels=['Name', 'Ticket', 'Fare'], axis=1), columns=['Sex','Embarked', 'Pclass'])#.drop(labels=['Embarked_C'], axis=1)
```

```python
# titanic_train['Embarked'] = titanic_train['Embarked'].fillna('S')
# titanic_train['Embarked'] = titanic_train['Embarked'].map({'S':0, 'C':1, 'Q':2})
titanic_train.head(10)
```

```python
# titanic_train = pd.get_dummies(titanic_train, columns=["Sex"])
# titanic_train.head()
```

```python
titanic_train = titanic_train.dropna(how='any')
titanic_train.head()
len(titanic_train)
```

```python
X_final = titanic_train.drop(["Survived"], axis=1)
y_final = titanic_train["Survived"].values.reshape(-1, 1)
print(X_final.shape, y_final.shape)
```

```python
len(y_final)
```

```python
y_final
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, random_state=1, stratify=y_final)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [50],
              'gamma': [0.0001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["Non-Survivors", "Survivors"]))
```
