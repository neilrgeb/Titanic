```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
import seaborn as sns
```

```python
titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')
```

```python
titanic_test.head(10)
```

```python
titanic_train.head(20)
```

```python
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train['Cabin'] = titanic_train['Cabin'].notnull().astype('int')
# X_train['Cabin'] = X_train['Cabin'].fillna(0)
titanic_train.head()
```

```python
titanic_train['Embarked'] = titanic_train['Embarked'].fillna('S')
titanic_train['Embarked'] = titanic_train['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
titanic_train = pd.get_dummies(titanic_train, columns=["Sex"])
titanic_train.head()
```

```python
titanic_train = titanic_train.dropna(how='any')
titanic_train.head()
len(titanic_train)
```

```python
X_final = titanic_train.drop(["Survived", "Ticket", "Name"], axis=1)
y_final = titanic_train["Survived"].values.reshape(-1, 1)
print(X_final.shape, y_final.shape)
```

```python
cor_map = titanic_train.drop(["Ticket", "Name", "PassengerId"], axis=1)
cor_map
```

```python
len(y_final)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, random_state=1, stratify=y_final)
```

```python
# X_final
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [10],
              'gamma': [0.0001]}
grid = GridSearchCV(model, param_grid, verbose=3)
grid.
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test)
```

```python
len(predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["Non-Survivors", "Survivors"]))
```

```python
import pickle

# Save to file in the current working directory
pkl_filename = "titanic_model_log.pkl"
with open(pkl_filename, 'wb') as file:
   pickle.dump(grid, file)

# Load from file
with open(pkl_filename, 'rb') as file:
   pickle_model = pickle.load(file)

# Calculate the accuracy score and predict target values
score = pickle_model.score(X_test, y_test)
print("Test score: {0:.2f} %".format(100 * score))
y_predict = pickle_model.predict(X_test)
```

```python
# Generate 4 clusters of random data.
from sklearn.datasets.samples_generator import make_blobs

data, _ = make_blobs(n_samples=300, centers=4,
                     cluster_std=0.60, random_state=0)
# Use n_clusters=4 as the k value
# We can see from the plot above that there are 4 clusters
from sklearn.cluster import KMeans, MeanShift, DBSCAN
k_means = KMeans()
from sklearn.model_selection import GridSearchCV

param_grid = { 'n_clusters': range(2, 30, 2) } 

grid_search_cv = GridSearchCV(k_means, param_grid, verbose=1)
grid_search_cv.fit(data)

import pandas as pd

print ('Best params are ', grid_search_cv.best_params_)

gridsearch_cv_results_df = pd.DataFrame({
    'n_clusters (k)': [params['n_clusters'] for params in grid_search_cv.cv_results_['params']],
      'split0_score':  grid_search_cv.cv_results_['split0_test_score'],
      'split1_score':  grid_search_cv.cv_results_['split1_test_score'],
      'split2_score':  grid_search_cv.cv_results_['split2_test_score']
     })

(gridsearch_cv_results_df.iloc[:,[0,1]]
    .plot(x='n_clusters (k)', y='split0_score'))

gridsearch_cv_results_df.iloc[:,[0,1]]
```
